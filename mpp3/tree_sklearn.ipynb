```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.tree import export_graphviz
import matplotlib.pyplot as plt
import cv2 as cv
%matplotlib inline
```

```python
iris = load_iris()
x = iris.data[:, 2:]
y = iris.target
x_names = iris.feature_names[2:]
y_names = iris.target_names
```

```python
tree_classifier = DecisionTreeClassifier(max_depth=3)
tree_classifier.fit(x,y)
```

```python
export_graphviz(tree_classifier, 'classifier.dot', feature_names = x_names,
               class_names=y_names)
! dot -Tpng classifier.dot -o classifier.png
img = cv.imread('classifier.png')
plt.figure(figsize = (20, 20))
plt.imshow(img)
```

```python
tree_regressor = DecisionTreeRegressor(max_depth=3)
tree_regressor.fit(x,y)
```

```python
export_graphviz(tree_regressor, 'regressor.dot', feature_names = x_names,
               class_names=y_names)
! dot -Tpng regressor.dot -o regressor.png
img = cv.imread('regressor.png')
plt.figure(figsize = (20, 20))
plt.imshow(img)
```
